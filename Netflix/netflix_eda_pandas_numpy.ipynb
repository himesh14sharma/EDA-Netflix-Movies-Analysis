# ðŸ”¹ Section 2: Data Cleaning

```python
# Replace all missing values in the director column with "Unknown".
df.loc[:, 'director'] = df['director'].fillna('Unknown')
```

```python
# Replace missing values in the country column with "Not Available".
df['country'].fillna("Not Available")
```

```python
# Remove rows where the rating column is missing.
df = df.dropna(subset =['rating'])
```

```python
# Check whether any missing values still exist after cleaning.
df.isnull().sum()
```

```python
# Convert the date_added column into datetime format.
df.rename(columns={"date_added": "datetime"})
```

# ðŸ”¹ Section 9: NumPy-based Numerical Analysis

```python
# Convert the release_year column into a NumPy array
years= df['release_year'].to_numpy()
```

```python
# Find the mean release year of Netflix content.
np.mean(years)
```

```python
# Find the minimum and maximum release year using NumPy.
np.max(years) , np.min(years)
```

```python
# Calculate the 25th, 50th, and 75th percentile of release years.
np.percentile(years,[25 , 50 , 75])
```

```python
# Compute the standard deviation of release years.
np.std(years)
```

# ðŸ”¹ Section 10: Advanced Filtering & Logical Analysis

```python
# Find all Movies released in India after 2017.
df[(df['type']=='Movie') & ( df['country'].str.contains('India' , na=False)) & (df['release_year']>2017)]
```

```python
# Find TV Shows added to Netflix after 2020.
df[(df['type'] == 'TV Show') & (df['date_added'].dt.year > 2020)]
```

```python
# Identify content where the duration is greater than 100 minutes
df[(df['type'] == 'Movie') & (df['duration'].str.contains('min', na=False)) &
    (pd.to_numeric(df['duration'].str.extract('(\d+)')[0], errors='coerce') > 100)]
```

```python
# Find titles that belong to both â€œDramasâ€ and â€œInternational Moviesâ€.
df[df['listed_in'].str.contains('Dramas', na=False) &
   df['listed_in'].str.contains('International', na=False)]
```

```python
# Find Netflix content released before the year 2000.
df[df['release_year']<2000]
```
